```python
#Prueba del CHI - cuadrado

#H0: no existe dependencia entre las variables
#H1: sis existe dependencia entre las variables

crosstab = pd.crosstab(data["Gender"], data["Purchased"])
from scipy.stats import chi2_contingency
chi2, p, dof, ex = chi2_contingency(crosstab)
print(chi2)
print(p)

#Si pvalur < 0.05, se rechaza la hipotesis  H0.
#Si pvalue > 0.05, no se rechaza la hipotesis H0
```
